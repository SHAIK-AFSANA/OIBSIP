```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## DATA COLLECTION AND PRE PROCESSING

```python
##Loading the data from CSV file to a pandas dataframe
raw_mail_data=pd.read_csv('/kaggle/input/mail-data/mail_data.csv')
print(raw_mail_data)
```

```python
# replace the null values with a null string
mail_data=raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
# printng the firast 5 rows of the data frame
mail_data.head()
```

```python
# checking the number of rows and columns
mail_data.shape
```

## LABEL ENCODING

```python
# label spam mail as 0; ham mail as 1;
mail_data.loc[mail_data['Category']=='spam','Category',]=0
mail_data.loc[mail_data['Category']=='ham','Category',]=1
```

```python
# separating the data as text and labels
X = mail_data['Message']
Y=mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

## SPLITTING THE DATA INTO TRAINING AND TESTING DATA

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
print(X.shape)
print(x_train.shape)
print(x_test.shape)
```

## FEATURE EXTRACTION

```python
# Transform the text data to feature vectors that can be used as input to he logistic regression
```

```python
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)

##convert y_train and y_test values as integers
y_train=y_train.astype('int')
y_test=y_test.astype('int')


```

```python
x_train_features=feature_extraction.fit_transform(x_train)
x_test_features=feature_extraction.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_train_features)
```

## TRAINING THE MODEL

```python
## Logistic Regression
```

```python
model=LogisticRegression()
## training the Logistic Regression model with the training data
model.fit(x_train_features,y_train)
```

## EVALUATING THE TRAINING MODEL

```python
## prediction on training data
prediction_on_training_data=model.predict(x_train_features)
accuracy_on_training_data=accuracy_score(y_train,prediction_on_training_data)
```

```python
print("Accuracy",accuracy_on_training_data)
```

```python
## prediction on training data
prediction_on_test_data=model.predict(x_test_features)
accuracy_on_test_data=accuracy_score(y_test,prediction_on_test_data)
```

```python
print("Acuuracy",accuracy_on_test_data)
```

## BUILDING A PREDICTIVE SYSTEM

```python
input_mail=["You got a cash prize of 1 lakh rupees."]
# convert text ot feature vectors
input_data_features=feature_extraction.transform(input_mail)
#making predictions
prediction=model.predict(input_data_features)
#print(prediction)
if prediction[0]==1:
    print("Ham Mail")
else:
    print("Spam Mail")
```
